```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Bank1.csv")
```

```python
dataset
```

```python
#making an arry of dataset
#iloc means the index location
#0 to 7
x=dataset.iloc[:,0:8].values
```

```python
x
```

```python
#create the Elbow chart
#wcss is the sum of the clusters
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,10):
  kmeans=KMeans(n_clusters=i,init='k-means++')
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
#find the number of clusters
#this will show the output in a graph format

plt.plot(range(1,10),wcss)
```

```python
#define the clusters for each record

kmeans1 = KMeans(n_clusters=3,init='k-means++')
y_kmeans = kmeans1.fit_predict(x)
```

```python
y_kmeans
```

```python
#Concatenate the dataset and the cluster index

pd.concat([dataset,pd.DataFrame(y_kmeans)],axis=1)
```

```python
#visualize the scatterpot

plt.scatter(x[y_kmeans==0,0],y=x[y_kmeans==0,1],s=50,c='red')
plt.scatter(x[y_kmeans==1,0],y=x[y_kmeans==1,1],s=50,c='blue')
plt.scatter(x[y_kmeans==2,0],y=x[y_kmeans==2,1],s=50,c='yellow')
```

```python
#Find the cluster centroid for each cluster

clusters = kmeans1.cluster_centers_
print(clusters)
```

```python
#Visualize the cluster centroids in the scatter plot
plt.scatter(x[y_kmeans==0,0],y=x[y_kmeans==0,1],s=50,c='red')
plt.scatter(x[y_kmeans==1,0],y=x[y_kmeans==1,1],s=50,c='blue')
plt.scatter(x[y_kmeans==2,0],y=x[y_kmeans==2,1],s=50,c='yellow')

plt.scatter(clusters[0][0],clusters[0][1],marker='*',s=200, color='black')
plt.scatter(clusters[1][0],clusters[1][1],marker='*',s=200, color='black')
plt.scatter(clusters[2][0],clusters[2][1],marker='*',s=200, color='black')
```

```python
#Separate the data frame into two parts

x=df.values
y=df['PEP']
```

```python
#here we deleted the pep

x=np.delete(x,7,axis=1)
```

```python
print(x)
```

```python
print(y)
```

```python
#Split the dataset into test and training set

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
#Find the prediction of the Using Decision tree


from sklearn import tree
df_clf = tree.DecisionTreeClassifier(max_depth=5)
df_clf.fit(x_train,y_train) # Corrected method name

df_clf.score(x_test,y_test)

y_pred = df_clf.predict(x_test)
df_clf.score(x_test,y_test)
```

```python
y_pred
```

```python
#Construct the confusion matrix

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```
